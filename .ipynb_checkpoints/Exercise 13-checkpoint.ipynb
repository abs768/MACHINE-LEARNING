```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.target
```

```python
df=pd.DataFrame(digits.data)
df['target']=digits.target
df
```

```python
X=df.drop('target',axis='columns')
y=df.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.4,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
model=GridSearchCV(KNeighborsClassifier(algorithm='auto'),{
    'n_neighbors':[1,2,3],
    'p':[2,3,4],
    'weights':['uniform','distance'],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute']   
},cv=5,return_train_score=False)
model.fit(train_X,train_y)
```

```python
df_cv=pd.DataFrame(model.cv_results_)
df_cv
```

```python
print(dir(model))
```

```python
model.best_score_
```

```python
model.best_params_
```

```python
model_new=KNeighborsClassifier(algorithm='auto',n_neighbors=1,p=3,weights='uniform')
```

```python
model_new.fit(train_X,train_y)
```

```python
model_new.score(test_X,test_y)
```

```python
y_predicted=model_new.predict(test_X)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(test_y,y_predicted)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
len(test_X)
```

```python
from sklearn.metrics import classification_report
report=classification_report(test_y,y_predicted)
print(report)
```
