```python
import pandas as pd
df=pd.read_csv("heart.csv")
df
```

```python
df.describe()
```

## Removing Outliers

```python
df[df['RestingBP']>df['RestingBP'].mean()+3*df['RestingBP'].std()]
```

```python
df1=df[df['RestingBP']<=df['RestingBP'].mean()+3*df['RestingBP'].std()]
```

```python
df1.shape
```

```python
df[df['Cholesterol']>df['Cholesterol'].mean()+3*df['Cholesterol'].std()]
```

```python
df2=df1[df1['Cholesterol']<=df1['Cholesterol'].mean()+3*df1['Cholesterol'].std()]
```

```python
df2.shape
```

```python
df[df['FastingBS']>df['FastingBS'].mean()+3*df['FastingBS'].std()]
```

```python
df[df['MaxHR']>df['MaxHR'].mean()+3*df['MaxHR'].std()]
```

```python
df[df['Oldpeak']>df['Oldpeak'].mean()+3*df['Oldpeak'].std()]
```

```python
df3=df2[df2['Oldpeak']<=df2['Oldpeak'].mean()+3*df2['Oldpeak'].std()]
```

```python
df3.shape
```

```python
df3
```

## Encoding Texts

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
```

```python
df3['ExerciseAngina']=df['ExerciseAngina'].apply(lambda x : 1 if x=='Y' else 0)
```

```python
df3.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)
```

```python
df3.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)
```

```python
df3
```

```python
df4=pd.get_dummies(df3,drop_first=True)
```

```python
df4
```

```python
X=df4.drop('HeartDisease',axis='columns')
X
```

```python
y=df4.HeartDisease
y
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

## Training

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=2)
```

## SVM

```python
from sklearn.svm import SVC
model1=SVC(C=1,kernel='rbf',gamma='auto')
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model2=LogisticRegression(solver='liblinear',multi_class='auto',C=5)
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(n_estimators=10)
model3.fit(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```
