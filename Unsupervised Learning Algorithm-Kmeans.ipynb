```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
df=pd.read_csv("income.csv")
df
```

```python
plt.scatter(df.Age,df['Income($)'],marker='+',color='red')
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
```

```python
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.figure(figsize=(5,3),dpi=130)
plt.scatter(df1.Age,df1['Income($)'],marker='+',color='red')
plt.scatter(df2.Age,df2['Income($)'],marker='*',color='blue')
plt.scatter(df3.Age,df3['Income($)'],marker='.',color='green')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df[['Income($)']]=scaler.transform(df[['Income($)']])
df
scaler.fit(df[['Age']])
df[['Age']]=scaler.transform(df[['Age']])
df
```

```python
model=KMeans()
```

```python
y_predicted=model.fit_predict(df[['Age','Income($)']])
```

```python
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
model.cluster_centers_
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.figure(figsize=(5,3),dpi=130)
plt.scatter(df1.Age,df1['Income($)'],marker='+',color='red',label='Income($)')
plt.scatter(df2.Age,df2['Income($)'],marker='*',color='blue',label='Income($)')
plt.scatter(df3.Age,df3['Income($)'],marker='.',color='green',label='Income($)')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='purple',marker='_',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

## Elbow Plot

```python
k_values=range(1,10)
sse=[]
for k in k_values:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_values,sse)
```
