```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("salaries.csv")
df
```

```python
features=df.drop('salary_more_then_100k',axis='columns')
label=df['salary_more_then_100k']
```

```python
features
```

```python
label
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
company_n=LabelEncoder()
job_n=LabelEncoder()
degree_n=LabelEncoder()
```

```python
features['company_n']=company_n.fit_transform(features['company'])
features['job_n']=job_n.fit_transform(features['job'])
features['degree_n']=degree_n.fit_transform(features['degree'])
```

```python
features
```

```python
features=features.drop(['company','job','degree'],axis='columns')
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,label,test_size=0.5)
```

```python
len(X_train)
```

```python
(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
