## Label Encoding and One-Hot-Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df3['RestingECG']=le.fit_transform(df3['RestingECG'])
```

```python
df3['ST_Slope'].unique()
```

```python
df4=df3.replace({
    'Up':1,
    'Flat':2,
    'Down':3
}
)
```

```python
df4['ST_Slope'].unique()
```

```python
df4['ExerciseAngina']=df4['ExerciseAngina'].apply(lambda x: 1 if(x=='Y') else 0)
```

```python
df4
```

```python
df4['Sex']=df4['Sex'].apply(lambda x: 1 if(x=='M') else 0)
```

```python
df4
```

```python
df5=pd.get_dummies(df4,drop_first=True)
```

```python
df5.head()
```

## Scaling

```python
X=df5.drop('HeartDisease',axis='columns')
X
```

```python
y=df5.HeartDisease
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

## Training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=3)
```

## Model SVM

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import cross_val_score
```

```python
k=cross_val_score(SVC(),X,y,cv=5)
k
```

```python
k.mean()
```

```python
model=SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model_1=BaggingClassifier(
    base_estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
```

```python
model_1.fit(X_train,y_train)
model_1.oob_score_
```

```python
model_1.score(X_test,y_test)
```

```python
l=cross_val_score(model_1,X,y,cv=5)
l
```

```python
l.mean()
```

## Model DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
m=cross_val_score(DecisionTreeClassifier(),X,y,cv=5)
m.mean()
```

```python
model_2=BaggingClassifier(
        base_estimator=DecisionTreeClassifier(),
        n_estimators=100,
        max_samples=0.8,
        oob_score=True,
        random_state=0
)
model_2.fit(X_train,y_train)
model_2.oob_score_
```

```python
n=cross_val_score(model_2,X,y,cv=5)
n.mean()
```
